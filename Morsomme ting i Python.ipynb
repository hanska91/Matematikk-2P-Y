```python
import time
import threading

def time_convert(sec):
    mins = sec // 60
    sec = sec % 60
    hours = mins // 60
    mins = mins % 60
    print("Time Lapsed = {0}:{1}:{2}".format(int(hours), int(mins), sec))

def stopwatch():
    input("Press Enter to start")
    start_time = time.time()
    stop_flag = threading.Event()
    
    def display_time():
        while not stop_flag.is_set():
            elapsed_time = time.time() - start_time
            mins = int(elapsed_time // 60)
            secs = int(elapsed_time % 60)
            hours = int(mins // 60)
            mins = int(mins % 60)
            print(f"\rTime: {hours:02}:{mins:02}:{secs:02}", end="")
            time.sleep(1)
    
    timer_thread = threading.Thread(target=display_time)
    timer_thread.daemon = True
    timer_thread.start()
    
    input("Press Enter to stop")
    stop_flag.set()
    end_time = time.time()
    time_lapsed = end_time - start_time
    time_convert(time_lapsed)

stopwatch()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Pendelvariabler
m = 2.5  # masse i kg
q = 10.0  # dempingsfaktor i Ns/m
g = 9.81  # tyngdeakselerasjon i m/s^2
l = 3.20  # lengde p친 pendelsnor i meter
y0 = 1.0  # initial posisjon i meter
y_prime0 = 0  # initial fart i meter/sekund

# Beregningsvariabler
a = m * l
b = q * l
c = m * g

# Tidsvariabler
dt = 0.01  # tidssteg
t = np.arange(0, 10, dt)  # tidsarray

# Differensialligning for pendelen
def pendel_derivater(state, t):
    y, y_prime = state
    y_double_prime = -b/a * y_prime - c/a * y
    return [y_prime, y_double_prime]

# Initialtilstand
init_state = [y0, y_prime0]

# L칮sning av differensialligningen
state = odeint(pendel_derivater, init_state, t)
y = state[:, 0]
y_prime = state[:, 1]
y_double_prime = -b/a * y_prime - c/a * y

# Beregning av analysene
omega = np.sqrt(g / l)  # vinkelhastighet
T = 2 * np.pi / omega  # svingetid/periodetid
f = omega / (2 * np.pi)  # frekvens
q_crit = 2 * m * omega  # kritisk dempingsfaktor

# Svingningstype
if q > q_crit:
    swing_type = "Overdempet svingning"
elif q < q_crit:
    swing_type = "Underdempet svingning"
else:
    swing_type = "Kritisk dempet svingning"

# Plotting av posisjonsgraf
plt.figure()
plt.plot(t, y)
plt.xlabel('Tid (s)')
plt.ylabel('Posisjon (m)')
plt.title('Posisjonsgraf for en matematisk pendel med demping')
plt.grid(True)
plt.show()

# Plotting av fartsgraf
plt.figure()
plt.plot(t, y_prime)
plt.xlabel('Tid (s)')
plt.ylabel('Fart (m/s)')
plt.title('Fartsgraf for en matematisk pendel med demping')
plt.grid(True)
plt.show()

# Plotting av akselerasjonsgraf
plt.figure()
plt.plot(t, y_double_prime)
plt.xlabel('Tid (s)')
plt.ylabel('Akselerasjon (m/s^2)')
plt.title('Akselerasjonsgraf for en matematisk pendel med demping')
plt.grid(True)
plt.show()

# Skriv ut analysene
print("Vinkelhastighet (픨): {:.2f} rad/s".format(omega))
print("Svingetid/periodetid (T): {:.2f} s".format(T))
print("Frekvens (f): {:.2f} Hz".format(f))
print()
print("Svingningstype:", swing_type)

# L칮sning av karakteristisk likning
r1 = ((-q*l) + np.sqrt(((q*l)**2)-(4*m*g*l*m))) / (2*m*l)
r2 = ((-q*l) - np.sqrt(((q*l)**2)-(4*m*g*l*m))) / (2*m*l)

# Beregning av A og B
A = (y_prime0 - r1 * y0) / (r2 - r1)
B = y0 - A

# Skriv ut funksjonsuttrykkene
print("Funksjonsuttrykk for posisjonsgrafen: y(t) = {:.2f} * e^({:.2f}t) + {:.2f} * e^({:.2f}t)".format(A, r1, B, r2))
print("Funksjonsuttrykk for fartsgrafen: v(t) = y'(t) = {:.2f} * e^({:.2f}t) + {:.2f} * e^({:.2f}t)".format(A * r1, r1, B * r2, r2))
print("Funksjonsuttrykk for akselerasjonsgrafen: a(t) = y''(t) = {:.2f} * e^({:.2f}t) + {:.2f} * e^({:.2f}t)".format(A * r1**2, r1, B * r2**2, r2))
```

```python
import pyautogui
import os
from PIL import ImageGrab
import tkinter as tk

# Create a folder to save the screenshots
folder_path = 'screenshots'
os.makedirs(folder_path, exist_ok=True)

# Function to select the region for the screenshot
def select_region():
    global x1, y1, x2, y2, rect
    root = tk.Tk()
    root.attributes("-alpha", 0.3)
    root.attributes("-fullscreen", True)
    canvas = tk.Canvas(root, cursor="cross")
    canvas.pack(fill=tk.BOTH, expand=True)
    rect = None
    root.bind("<Button-1>", lambda event: on_click(event, canvas))
    root.bind("<B1-Motion>", lambda event: on_drag(event, canvas))
    root.bind("<ButtonRelease-1>", lambda event: on_release(event, root))
    root.mainloop()

def on_click(event, canvas):
    global x1, y1, rect
    x1, y1 = event.x, event.y
    rect = canvas.create_rectangle(x1, y1, x1, y1, outline='red', fill='gray', stipple='gray50')

def on_drag(event, canvas):
    global x2, y2, rect
    x2, y2 = event.x, event.y
    canvas.coords(rect, x1, y1, x2, y2)

def on_release(event, root):
    global x2, y2
    x2, y2 = event.x, event.y
    root.quit()
    root.destroy()

# Initialize coordinates
x1 = y1 = x2 = y2 = 0

# Select the region
select_region()
region = (min(x1, x2), min(y1, y2), abs(x2 - x1), abs(y2 - y1))

# Ensure the region is within the screen bounds
screen_width, screen_height = pyautogui.size()
region = (
    max(0, min(region[0], screen_width)),
    max(0, min(region[1], screen_height)),
    max(0, min(region[2], screen_width - region[0])),
    max(0, min(region[3], screen_height - region[1]))
)

# Take a screenshot using the selected region
ss = pyautogui.screenshot(region=region)
ss.save(os.path.join(folder_path, 'Bilde1.png'))

# Take a full screenshot
ss = ImageGrab.grab()
ss.save(os.path.join(folder_path, 'Bilde2.png'))

print(f"Screenshots saved in the '{folder_path}' folder.")
```

```python
def print_triangle(height):
    for i in range(1, height + 1):
        print(' ' * (height - i) + '*' * (2 * i - 1))

def print_square(size):
    for i in range(size):
        print('*' * size)

def print_rectangle(width, height):
    for i in range(height):
        print('*' * width)

def print_diamond(height):
    for i in range(1, height + 1):
        print(' ' * (height - i) + '*' * (2 * i - 1))
    for i in range(height - 1, 0, -1):
        print(' ' * (height - i) + '*' * (2 * i - 1))

def print_circle(radius):
    for y in range(-radius, radius + 1):
        for x in range(-radius, radius + 1):
            if x**2 + y**2 <= radius**2:
                print('*', end='')
            else:
                print(' ', end='')
        print()

# Example usage
print("Triangle:")
print_triangle(5)
print("\nSquare:")
print_square(5)
print("\nRectangle:")
print_rectangle(7, 4)
print("\nDiamond:")
print_diamond(5)
print("\nCircle:")
print_circle(5)
```
